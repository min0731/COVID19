# 기본제공데이터 전처리

```python
import pandas as pd
import numpy as np
import os
from datetime import datetime, timedelta
from collections import Counter

os.chdir("C:/Users/dhy03/Desktop/코로나공모전200925/기본 제공 데이터")
```

```python
base_dataset=pd.read_csv('full_data_0923.csv')
base_dataset.columns
```

```python
sub_dat=base_dataset[['확진자', '성별','확진일자', '증상발현일','연령대', '무증상/조사중',
                     '경기번호','재검출', '감염경로','구분']]
sub_dat['확진일자'] = pd.to_datetime(sub_dat['확진일자'], format="%Y. %m. %d.")
sub_dat['증상발현일'] = pd.to_datetime(sub_dat['증상발현일'], format="%Y. %m. %d.")
sub_dat['연령대'] = sub_dat['연령대'].astype(str)+'대'
sub_dat.info()
```

```python
sub_dat['무증상/조사중'].value_counts()

무증상여부=[]
for i in range(0,len(sub_dat)):
    if sub_dat.loc[i,'무증상/조사중']=='무증상':
        무증상여부.append('무증상O')
    else:
        무증상여부.append('무증상X')
        
Counter(무증상여부)
```

```python
sub_dat['재검출'].value_counts()
재검출여부=[]
for i in range(0,len(sub_dat)):
    cell_=sub_dat.loc[i,'재검출']
    if '재' in cell_:
        재검출여부.append('재검출O')
    else:
        재검출여부.append('재검출X')
        
Counter(재검출여부)
```

```python
sub_dat['감염경로'].value_counts()
해외감염여부=[]
for i in range(0,len(sub_dat)):
    if sub_dat.loc[i,'감염경로']=='해외유입':
        해외감염여부.append('해외유입O')
    else:
        해외감염여부.append('해외유입X')

Counter(해외감염여부)
```

```python

Primary_접촉자=[]
primary_list=['Primary','primary','piramary','primary?','Primary가능성 있음','Primary(추정)','Primary?']
접촉자_list=['접촉자','접촉자?']

for i in range(0,len(sub_dat)):
    cell_=sub_dat.loc[i,'구분']
    if cell_ in primary_list:
        Primary_접촉자.append('Primary')
    elif cell_ in 접촉자_list:
        Primary_접촉자.append('접촉자')
    else:
        Primary_접촉자.append('알수없음')
        
Counter(Primary_접촉자)
```

```python
경기번호_split=sub_dat['경기번호'].str.split('-',n=1,expand=True)
경기지역=경기번호_split[0]
경기지역
```

```python
증상발현후_확진기간=sub_dat['확진일자']-sub_dat['증상발현일']
증상발현후_확진기간_fin=[]

for i in range(0,len(증상발현후_확진기간)):
    cell_=증상발현후_확진기간[i]
    증상발현후_확진기간_fin.append(cell_.days)
```

```python
new_df=pd.DataFrame(list(zip(증상발현후_확진기간_fin,경기지역,Primary_접촉자,
                            해외감염여부,무증상여부,재검출여부)),
                   columns=['증상발현후_확진기간','경기지역','Primary_접촉자',
                            '해외감염여부','무증상여부','재검출여부'])
new1_df=sub_dat[['확진일자','성별','연령대']]

fin_df=pd.concat([new1_df,new_df],axis=1)
fin_df=fin_df.sort_values(by='확진일자')
fin_df=fin_df.reset_index(drop=True)
fin_df
```

```python
확진일자_df=fin_df['확진일자'].value_counts()
확진일자_df=pd.DataFrame(확진일자_df)
확진일자_df=확진일자_df.reset_index(drop=False)
확진일자_df=확진일자_df.rename(columns={'index':'확진일자','확진일자':'확진자수'},inplace = False)
확진일자_df=확진일자_df.sort_values(by='확진일자')
확진일자기준_df=확진일자_df.reset_index(drop=True)
확진일자기준_df
```

```python
drange = pd.date_range(start = 확진일자기준_df.iloc[0,0], end = 확진일자기준_df.iloc[-1,0])
datefull_df=pd.DataFrame(drange, columns=['확진일자'])

datefull_dffin=pd.merge(datefull_df,확진일자기준_df,how='left',on='확진일자')
datefull_dffin=datefull_dffin.fillna(0)
datefull_dffin
```

```python
fixed_df=pd.DataFrame()
for i in range(0,len(확진일자기준_df)):
    date_=확진일자기준_df.iloc[i,0]
    fixed_df.loc[i,'확진일자']=date_
    총확진자수=확진일자기준_df.iloc[i,1]
    fixed_df.loc[i,'총확진자수']=총확진자수
    
    sub_df=fin_df[fin_df.확진일자==date_]
    
    ##성별 관련 변수
    mw=sub_df['성별']
    fixed_df.loc[i,'남성확진자수']=np.sum(mw=='M')
    fixed_df.loc[i,'여성확진자수']=np.sum(mw=='F')
    
    ##증상발현후_확진기간
    tt=sub_df['증상발현후_확진기간']
    tt_num=tt.isna().sum()
    fixed_df.loc[i,'증상발현후_확진기간']=tt[~np.isnan(tt)].mean()
    fixed_df.loc[i,'증상발현_데이터수']=총확진자수-tt_num
    
    ##Primary_접촉자
    ss=sub_df['Primary_접촉자']
    fixed_df.loc[i,'Primary수']=np.sum(ss=='Primary')
    fixed_df.loc[i,'접촉자수']=np.sum(ss=='접촉자')
    
    ##해외감염여부
    rr=sub_df['해외감염여부']
    fixed_df.loc[i,'해외유입자수']=np.sum(rr=='해외유입O')
    
    ##무증상여부
    rr=sub_df['무증상여부']
    fixed_df.loc[i,'무증상확진자수']=np.sum(rr=='무증상O')
    
    ##재검출여부
    rr=sub_df['재검출여부']
    fixed_df.loc[i,'재검출확진자수']=np.sum(rr=='재검출O')
    
os.chdir("C:/Users/dhy03/Desktop/코로나공모전200925")

#fixed_df.to_csv('기본데이터확진자정리.csv',encoding="cp949",index=False)
fixed_df
```
